```python
def get_last_price(ticker: str) -> float | None:
    url = f"https://iss.moex.com/iss/engines/stock/markets/shares/boards/TQBR/securities/{ticker}.json"
    params = {"iss.meta": "off", "iss.only": "marketdata"}

    r = requests.get(url, params=params, timeout=20).json()

    md = r.get("marketdata", {})
    columns = md.get("columns", [])
    data = md.get("data", [])

    if not data or not columns:
        return None

    row = data[0]

    # Порядок предпочтения цен (на практике часто LAST=None, а есть LCLOSEPRICE/PREVPRICE)
    candidates = ["LAST", "MARKETPRICE", "LCLOSEPRICE", "PREVPRICE"]
    for col in candidates:
        if col in columns:
            v = row[columns.index(col)]
            if v is not None:
                try:
                    return float(v)
                except (TypeError, ValueError):
                    pass

    return None
```

```python
def get_shares_outstanding_moex(ticker: str) -> float | None:
    """
    Пытается получить количество акций (shares outstanding) из MOEX ISS.
    Возвращает None, если не найдено.
    """
    url = f"https://iss.moex.com/iss/securities/{ticker}.json"
    params = {"iss.meta": "off", "iss.only": "description"}

    r = requests.get(url, params=params, timeout=20).json()
    desc = r.get("description", {})
    data = desc.get("data", [])
    if not data:
        return None

    # В description строки обычно: [name, title, value, type, sort_order, is_hidden, precision]
    # Мы берём value (index 2)
    candidates = {"ISSUESIZE", "ISSUESIZEPLACED"}  # второе иногда встречается вместо первого

    found = {}
    for row in data:
        if len(row) >= 3 and row[0] in candidates:
            try:
                val = float(str(row[2]).replace(" ", "").replace(",", "."))
                found[row[0]] = val
            except (TypeError, ValueError):
                pass

    # Приоритет: ISSUESIZE, потом ISSUESIZEPLACED
    if "ISSUESIZE" in found:
        return found["ISSUESIZE"]
    if "ISSUESIZEPLACED" in found:
        return found["ISSUESIZEPLACED"]

    return None
```

```python
def get_shares_outstanding_smartlab(ticker: str, session: requests.Session | None = None) -> float | None:
    s = session or requests.Session()
    s.headers.update({
        "User-Agent": "Mozilla/5.0",
        "Accept-Language": "ru-RU,ru;q=0.9,en;q=0.8",
    })

    # обычно на странице /q/<ticker>/ есть блок с параметрами
    url = f"https://smart-lab.ru/q/{ticker}/"
    try:
        r = s.get(url, timeout=20)
    except requests.RequestException:
        return None

    if r.status_code != 200 or not r.text:
        return None

    soup = BeautifulSoup(r.text, "lxml")
    text = soup.get_text(" ", strip=True).lower()

    # эвристика: "акций в обращении 21.6 млрд" / "shares outstanding"
    # Парсинг стабильным быть не может, но часто работает.
    m = re.search(r"(акц(ий|ии)\s+в\s+обращении|shares\s+outstanding)[^\d\-]*([\d\.,]+)\s*(млрд|млн)?", text)
    if not m:
        return None

    num = m.group(3).replace(",", ".")
    try:
        val = float(num)
    except ValueError:
        return None

    unit = m.group(4)
    if unit == "млрд":
        val *= 1_000_000_000
    elif unit == "млн":
        val *= 1_000_000

    return val
```

```python
def get_shares_outstanding(ticker: str) -> float | None:
    v = get_shares_outstanding_moex(ticker)
    if v is not None and v > 0:
        return v

    # fallback 
    v = get_shares_outstanding_smartlab(ticker)
    if v is not None and v > 0:
        return v

    return None
```

```python
def _to_number_ru(x: str) -> float | None:
    """
    Преобразует строку вида '1 234,56' / '1,234.56' / '—' в float.
    """
    if x is None:
        return None
    s = str(x).strip()
    if not s or s in {"—", "-", "n/a", "NA", "None"}:
        return None

    # убираем неразрывные пробелы и обычные
    s = s.replace("\xa0", " ").replace(" ", "")

    # если встречается и '.' и ',', пытаемся угадать формат
    # RU формат: 1234,56
    # EN формат: 1,234.56
    if "," in s and "." in s:
        # если точка после запятой -> вероятно 1,234.56
        if s.rfind(".") > s.rfind(","):
            s = s.replace(",", "")
        else:
            s = s.replace(".", "").replace(",", ".")
    else:
        # если только запятая — считаем десятичным разделителем
        if "," in s:
            s = s.replace(",", ".")

    # оставляем только цифры, минус и точку
    s = re.sub(r"[^0-9\.\-]", "", s)
    try:
        return float(s)
    except ValueError:
        return None
```

```python
def get_net_income_smartlab(ticker: str, session: requests.Session | None = None) -> float | None:
    """
    Пытается получить годовую чистую прибыль (net income) для компании по тикеру (MOEX),
    используя Smart-Lab.

    Возвращает значение в РУБЛЯХ, если удаётся распарсить.
    Если не удалось — возвращает None.

    Важно: Smart-Lab может показывать данные в млн/млрд — это надо контролировать
    по заголовку/единицам на странице. Ниже есть обработка "млн/млрд" как эвристика.
    """
    s = session or requests.Session()
    s.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                      "(KHTML, like Gecko) Chrome/120.0 Safari/537.36",
        "Accept-Language": "ru-RU,ru;q=0.9,en;q=0.8",
    })

    # 1) Попытка: если у Smart-Lab есть JSON (не гарантируется)
    # Некоторые сайты имеют endpoint вида /q/<ticker>/f/y/ в JSON — но это нестабильно.
    # Поэтому пробуем мягко, без ожиданий.
    possible_json_urls = [
        f"https://smart-lab.ru/q/{ticker}/f/y/?output=json",
        f"https://smart-lab.ru/q/{ticker}/f/y/?format=json",
    ]

    for url in possible_json_urls:
        try:
            r = s.get(url, timeout=20)
            if r.status_code == 200 and "json" in (r.headers.get("Content-Type", "").lower()):
                data = r.json()
                # Структура может отличаться. Попробуем найти ключи по смыслу.
                # Ищем что-то похожее на net profit / чистая прибыль.
                # Если найдём — вернём самое свежее.
                # (Если сработает — круто; если нет — идём в HTML)
                if isinstance(data, dict):
                    # эвристика: ищем list of rows
                    rows = data.get("data") or data.get("rows") or data.get("items")
                    if isinstance(rows, list):
                        # ищем колонку "Чистая прибыль" / "Net profit"
                        for row in rows[::-1]:
                            # row может быть dict
                            if isinstance(row, dict):
                                for k in row.keys():
                                    lk = k.lower()
                                    if "чист" in lk and "приб" in lk or "net" in lk and "profit" in lk:
                                        val = _to_number_ru(row.get(k))
                                        if val is not None:
                                            return val
        except Exception:
            pass  # не получилось — не страшно

    # 2) Надёжнее: HTML страница финансовых показателей (годовые)
    # Типичный путь у Smart-Lab для финансов: /q/<TICKER>/f/y/
    url = f"https://smart-lab.ru/q/{ticker}/f/y/"

    try:
        r = s.get(url, timeout=20)
    except requests.RequestException:
        return None

    if r.status_code != 200 or not r.text:
        return None

    soup = BeautifulSoup(r.text, "lxml")

    # На странице обычно есть таблица. Ищем строку, где в первом столбце "Чистая прибыль"
    tables = soup.find_all("table")
    if not tables:
        return None

    # эвристика множителя (млн/млрд) — иногда указано в заголовке рядом
    page_text = soup.get_text(" ", strip=True).lower()
    multiplier = 1.0
    if "млрд" in page_text:
        multiplier = 1_000_000_000.0
    elif "млн" in page_text:
        multiplier = 1_000_000.0

    def row_is_net_income(label: str) -> bool:
        l = label.strip().lower()
        # варианты подписи на рус/англ
        return (
            "чистая прибыль" in l
            or "чист.приб" in l
            or "net profit" in l
            or ("прибыль" in l and "чист" in l)
        )

    # ищем самую подходящую таблицу
    for table in tables:
        rows = table.find_all("tr")
        for tr in rows:
            tds = tr.find_all(["td", "th"])
            if len(tds) < 2:
                continue

            label = tds[0].get_text(" ", strip=True)
            if not label:
                continue

            if row_is_net_income(label):
                # Берём последнее (самое правое) значение в строке
                values = [td.get_text(" ", strip=True) for td in tds[1:] if td.get_text(strip=True)]
                if not values:
                    return None

                last_val = values[-1]
                num = _to_number_ru(last_val)
                if num is None:
                    return None

                # применяем множитель, если определили "млн/млрд"
                return float(num) * multiplier

    return None
```

```python
def calculate_pe(ticker: str) -> dict:
    price = get_last_price(ticker)
    shares = get_shares_outstanding(ticker)
    if price is None or shares is None:
        return -1.0
    
    net_income = get_net_income_smartlab(ticker)
    if net_income is None:
        return -1.0
    
    eps = net_income / shares
    pe = price / eps if eps > 0 else -1.0

    return pe
```

```python
companies = {}
with Client(t_invest_token) as client:
    
    r = client.instruments.shares()
    for share in r.instruments:
        if share.class_code == "TQBR"and share.currency == "rub" and share.isin is not None:
            PE = None
            PE = calculate_pe(share.ticker)
            if PE is None:
                PE = -1
            
            companies[share.ticker] = { 
                "share": share, 
                "dividends": [],
                "filtered_dividends": [],
                "PE": float(PE)
            }

print("Загружено ", len(companies.keys()), " компаний")
print(", ".join((str(element)+"_PE:"+str(companies[element]["PE"])) for element in companies.keys()))
```

```python
with Client(t_invest_token) as client:

    for ticker in companies.keys():
        r = client.instruments.get_dividends(figi=companies[ticker]["share"].figi, from_=datetime.datetime(2000, 1, 1))
        for dividend in r.dividends:
            companies[ticker]["dividends"] += [dividend]
            value = (dividend.yield_value.units + dividend.yield_value.nano*1.0e-9)
            
            date = dividend.last_buy_date
            if(date > datetime.datetime.now(pytz.utc) ):
                date =  datetime.datetime(
                    datetime.datetime.today().year, 
                    datetime.datetime.today().month, 
                    datetime.datetime.today().day, 0, 0, 0, tzinfo=pytz.utc)
            
            if (     len(companies[ticker]["filtered_dividends"]) > 0
                # укрупняем выплаты до квартала 
                and  companies[ticker]["filtered_dividends"][len(companies[ticker]["filtered_dividends"]) -1]["date"] 
                        < date + datetime.timedelta(days=89)
               ):
                companies[ticker]["filtered_dividends"][len(companies[ticker]["filtered_dividends"]) -1]["value"] += value
            else:
                companies[ticker]["filtered_dividends"] += [{"date": date, "value": value}]



print("Загружено дивидендов ", len(companies.keys()), " компаний")
print(", ".join((f"{str(element)}_Divs:{str(len(companies[element]["filtered_dividends"]))}") for element in companies.keys()))
    
```

# Обработка данных

```python
#отсекаем компании не платившие дивиденты больше года или  менее однго раза в год
filtred_companies = {}
excluded = []
for ticker in companies.keys():
    if(  
            len(companies[ticker]["filtered_dividends"]) >= 2
        and companies[ticker]["filtered_dividends"][0]["date"] + datetime.timedelta(days=365) >= datetime.datetime.now(pytz.utc) 
        and len([d for d in companies[ticker]["dividends"] if d.dividend_net.currency != "rub"]) == 0
      ): 
        filtred_companies[ticker] = companies[ticker]
    else:
        excluded += [ticker]

print("Исключенные из анализа компании: ", len(excluded))
print(", ".join(str(ticker) for ticker in excluded))
print("")
print("Оставшиеся компании: ", len(filtred_companies.keys()) )
print(", ".join(str(ticker) for ticker in filtred_companies.keys()))
```

## Лучшие компании

```python
# строим серии графиков

def calck_graphs( input_companies, years_limit ):
    result = []
    for ticker in input_companies.keys():

        data = [ {"date": d["date"], "value": d["value"], "year_value":0 } for d in input_companies[ticker]["filtered_dividends"] ] 
        graph = {
            "average": 0, 
            "x":[], 
            "y":[], 
            "ticker": input_companies[ticker]["share"].ticker,
            "name": input_companies[ticker]["share"].name,
            "sector": input_companies[ticker]["share"].sector,
            "data": data,
            "PE": input_companies[ticker]["PE"]
        }
        
        sum_value = 0
        sum_days = 0
        for div_id in range(len(data)-2, -1, -1):
            date = data[div_id]["date"]
            
            #отбираем данные за последние years_limit года
            if date + datetime.timedelta(days=365*years_limit) < datetime.datetime.now(pytz.utc):
                continue
            
            #определяем количество дней с последней выплаты дивидендов
            days = (date - data[div_id+1]["date"]).days
            
            #определяем среднее годовое значение дивидендов за дни после предыдущей выплаты
            data[div_id]["year_value"] = data[div_id]["value"] / ( days / 365 ) 
            
            graph["x"] += [date]
            graph["y"] += [data[div_id]["year_value"]]
            
            sum_value += data[div_id]["year_value"] * days
            sum_days += days
    
        graph["average"] = sum_value / sum_days
        result += [graph]
    
    return sorted(result, key=lambda fc: fc["average"], reverse=True)
```

```python
#указываем количество последних лет, за которые будет произведен анализ
graphs = calck_graphs(filtred_companies, 2.5)
```

```python
legend_colors = {
    "materials": "lightblue",
    "consumer": "pink",
    "industrials": "grey",
    "it": "yellow",
    "energy": "black",
    "utilities": "orange",
    "telecom": "red",
    "financial": "green",
    "health_care": "lightgreen",
    "real_estate": "brown",
    "other": "blue"
}
```

```python
x = []
y = []
colors = []
patchs = {}
for g in graphs:
    
    x += [g["name"] + " (" + g["ticker"] + ")"]
    y += [g["average"]]
    
    sector = g["sector"]
    colors += [legend_colors[sector]]
    patchs[sector] = mpatches.Patch(color=legend_colors[sector], label=sector)

fig, ax = plt.subplots(figsize=(20,10))
p = ax.bar(x, y, color=colors)
ax.bar_label(p, fmt="%.1f", rotation='vertical', padding = 5)
ax.tick_params(axis='x', labelrotation=90)
ax.legend(handles=patchs.values())
```

```python
plt.figure(figsize=(20,10))
for g in graphs:

    if g["PE"] <= 0 or g["PE"] > 20:
        continue
    
    plt.scatter(
        g["PE"],
        g["average"],
        color=legend_colors[g["sector"]],
        s=100,
        alpha=0.8,
        edgecolors="black"
    )
    # Подпись точки
    plt.text(
        float(g["PE"]) + 0.3,
        g["average"] + 0.15,
        g["ticker"],
        fontsize=9
    )

# Оформление
plt.xlabel("P/E")
plt.ylabel("Средние годовые дивиденды, %")
plt.title("P/E vs Dividend Yield")

plt.gca().invert_xaxis()
plt.show()
```

```python
def draw_graphs(gr, start, end):
    plt.figure(figsize=(20,20))
    legend = []
    for i in range(start,end):
        if i >= len(gr):
            break
            
        fc = gr[i]
        legend += [fc["name"]]
        plt.plot(fc["x"], fc["y"])
    plt.legend(legend)
```
